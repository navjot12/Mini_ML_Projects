```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
%matplotlib inline

from sklearn.decomposition import PCA
```

```python
ds = pd.read_csv('./titanic.csv')
ds.head(n=10)
```

```python
ds.columns
#columns that matter are: Pclass, Sex, Age, SibSp, Parch, Fare, Embarked
```

```python
ds.describe()
```

```python
ds['Age'] = ds['Age'].fillna(ds['Age'].mean())
ds['Sex'] = ds['Sex'].replace('female', 1)
ds['Sex'] = ds['Sex'].replace('male', 0)
ds['Embarked'] = ds['Embarked'].replace('C', 1)
ds['Embarked'] = ds['Embarked'].replace('Q', 2)
ds['Embarked'] = ds['Embarked'].replace('S', 3)

ds = ds.drop(labels = ['Name', 'Ticket', 'Cabin', 'PassengerId'], axis = 1)
```

```python
ds = ds.dropna()
ds.describe()
```

```python
data = np.asarray([ds['Age'], ds['Sex'], ds['Pclass'], ds['Parch'], ds['SibSp'], ds['Fare'], ds['Embarked'], ds['Survived']]).astype('float').T
Y = ds['Survived'].values
cols = ['Age', 'Sex', 'Pclass', 'Parch', 'SibSp', 'Fare', 'Embarked', 'Survived']
```

```python
pca = PCA(n_components=8)
X = pca.fit_transform(data)
v = pca.explained_variance_ratio
```
